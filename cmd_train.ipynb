```python
import bert_score
print(bert_score.__version__)
```

```python
with open("save/YAGO43kET/ET_test_outputs_test.txt", "r", encoding="utf-8") as f:
    predictions = f.readlines()
with open("save/YAGO43kET/ET_test_targets_test.txt", "r", encoding="utf-8") as f:
    targets = f.readlines()
```

```python
from bert_score import score
import numpy as np
def calc_bert_score(targets, predictions):
    total_score_p = 0
    total_score_r = 0
    total_score_f1 = 0
    Precision, Recall, F1 = score(predictions, targets, lang="en", verbose=True)
    for p, r, f1 in zip(Precision.numpy().tolist(), Recall.numpy().tolist(), F1.numpy().tolist()):
        total_score_p += p
        total_score_r += r
        total_score_f1 += f1
    score_p = total_score_p/len(Precision)
    score_r = total_score_r/len(Precision)
    score_f1 = total_score_f1/len(Precision)
    print(score_p, score_r, score_f1)
```

```python
calc_bert_score(targets, predictions)
```

```python
Precision, Recall, F1 = score(predictions, targets, lang="en", verbose=True)
```

```python
score_p = Precision.sum()/len(Precision)
score_r = Recall.sum()/len(Recall)
score_f1 = F1.sum()/len(F1)
```
